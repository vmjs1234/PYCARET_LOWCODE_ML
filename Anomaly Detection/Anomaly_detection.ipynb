```python
import pandas as pd
### load sample dataset from pycaret dataset module
from pycaret.datasets import get_data
from google.colab import files
uploaded_files = files.upload()
data = pd.read_csv('Train_data.csv')
print(f"Successfully loaded {len(data)} rows.")
```

```python
# import pycaret anomaly and init setup
from pycaret.anomaly import *
s = setup(data, session_id = 123,use_gpu=True)
```

```python
# import AnomalyExperiment and init the class
from pycaret.anomaly import AnomalyExperiment
exp = AnomalyExperiment()
```

```python
# check the type of exp
type(exp)
```

```python
# init setup on exp
exp.setup(data, session_id = 123)
```

```python
# train iforest model
iforest = create_model('iforest')
iforest
```

```python
# to check all the available models
models()
```

```python
iforest_anomalies = assign_model(iforest)
iforest_anomalies
```

```python
# Make predictions on the test data
predictions = predict_model(iforest, data=data)
predictions[predictions['Anomaly'] == 1].head(10) # printing few anamolies
```

```python

# The 'Anomaly' column will contain 1 for anomalies and 0 for normal instances
# The 'Anomaly_Score' column will contain the anomaly scores

# Print summary of anomalies
anomaly_count = predictions['Anomaly'].sum()
print(f"Number of anomalies detected: {anomaly_count}")
print(f"Percentage of anomalies: {(anomaly_count / len(predictions)) * 100:.2f}%")
```
