```python
## Importing necessary libraries
import pandas as pd
import numpy as np
from pycaret.arules import *
```

```python
from google.colab import files
import io
uploaded_file = files.upload()
data = pd.read_csv(io.BytesIO(uploaded_file["Groceries_dataset.csv"]))
data.head()
df = pd.read_csv('Groceries_dataset.csv')
df.dropna(inplace = True)
df.head()
```

```python
setup(data, transaction_id = "Member_number", item_id = "itemDescription")
```

```python
association_rules = create_model(metric = "confidence", threshold = 0.5, min_support = 0.04)
association_rules.head()
```

```python
# @title consequents vs antecedent support

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(association_rules['consequents'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(association_rules, x='antecedent support', y='consequents', inner='stick', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
# @title support vs confidence

from matplotlib import pyplot as plt
association_rules.plot(kind='scatter', x='support', y='confidence', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```
